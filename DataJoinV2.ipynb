```python
import pandas as pd
import numpy as np
```

```python
#import cleaned data file 1 and drop last row (it's empty)
df1 = pd.read_csv('data1_cleaned.csv')
df1 = df1.drop([144])
df1.Month=df1.Month.astype(int)
df1 = df1.rename(columns={'Depature Delay Time': 'dep_delay'})
df1.dep_delay=df1.dep_delay.astype(int)
print('Any empty Values:\n',df1[df1.isnull().any(axis=1)])
df1
```

```python
#import data set 2
df2 = pd.read_csv('Data2.txt', sep = '|')
df2
```

```python
#import dataset 3
#**************** IMPORTANT: SOUTHWEST IS MISSING FLIGHT DATA FROM ORD ***************

df_nulls = pd.read_csv('Data3.csv', skiprows=3)
print('Rows that contain null values: \n',df_nulls[df_nulls.isnull().any(axis=1)])
print('\nRows that are from Chicago: \n',df_nulls.loc[df_nulls.Airport == 'ORD'])

#DROP ROWS WITH EMPTY VALUES
df3 = df_nulls.dropna()
df3.No_of_Flights=df3.No_of_Flights.astype(int)

print('Rows that contain empty values:\n', df3[df3.isna().any(axis=1)])

df3
```

```python
#replace carrierID values on df3 with Carrier name from dataset2

df4 = df3.merge(df2, on='CarrierID', how='left')
print('Rows that contain null values: \n',df4[df4.isnull().any(axis=1)])
df4
```

```python
#Find those pesky whitespaces making the data values different on each dataset

print(df1.Airport.unique())
print('Airport:',df4.Airport.unique() == df1.Airport.unique())
print('month: ',df4.Month.unique() == df1.Month.unique())
print('Carrier: ',df4.Carrier.unique() == df1.Carrier.unique())

print(df4.Carrier.unique())

df1.Carrier = df1.Carrier.str.rstrip()
df4.Carrier = df4.Carrier.str.rstrip()

print('Carrier: ',df4.Carrier.unique() == df1.Carrier.unique())
```

```python
#add dep_delay Series from df1 to df4

#rename df1.Airline to Carrier
df1 = df1.rename(columns={'Airline':'Carrier'})
print(df1)
print(df4)

df_all = pd.merge(df1,df4, 
                  left_on=['Airport', 'Month', 'Carrier', 'City'], 
                  right_on=['Airport', 'Month', 'Carrier', 'City'], 
                  how='inner')


print('Rows with null values:\n',df_all[df_all.isnull().any(axis=1)])
df_all = df_all.dropna()
```

```python
#Write the joined dataset to a file

df_all.to_csv('JoinedData_csv')

df_all.to_excel('JoinedData_excel.xlsx')

df_nulls[df_nulls.isnull().any(axis=1)].to_excel('Missing_Data.xlsx')
```
